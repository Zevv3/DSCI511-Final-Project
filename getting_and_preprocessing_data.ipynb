```python
def get_details(url, token):
  headers = {
              'Authorization': f'Bearer {token}'
          }
  response = requests.get(url, headers=headers)
  details_data = response.json()['data']
  return details_data
```

```python
# dates = ['2018-01','2018-02','2018-03', '2018-04', '2018-05', '2018-06', '2018-07', '2018-08', '2018-09', '2018-10', '2018-11', '2018-12']
city_codes = ['NYC', 'MIA', 'LAX', 'SEA', 'CHI', 'MEX', 'YTO', 'YVR', 'SAO', 'RIO', 'BOG', 'TYO', 'DEL', 'TLV', 'PAR', 'MAD', 'MOW', 'LON', 'SYD']

def get_data(city_codes, departure_date, access_token):
    url = 'https://test.api.amadeus.com/v1/shopping/flight-destinations'
    headers = {
            'Authorization': f'Bearer {access_token}'
        }
    all_data = {}
    for city in city_codes:
        new_token = get_token()
        print(f'now looking at {city}')
        city_data = {}
        params = {
            'origin': city,
            'departureDate': departure_date,
        }
        response = requests.get(url, headers=headers, params=params)
        if response.status_code == 200:
            data = response.json()['data']
            count = 0
            for entry in data:
                count += 1
                print(f'{city} entry {count}')
                # time.sleep(2)
                details_url = entry['links']['flightOffers']
                details_data = get_details(details_url, new_token)
                if details_data:
                  details_data = details_data[0]
                  currency = details_data['price']['currency']
                  seats = details_data['numberOfBookableSeats']

                  # outbound flight data gathering
                  outbound_flight = details_data['itineraries'][0]
                  outbound_flight_duration = 'No data'
                  outbound_flight_terminal = ''
                  try:
                    outbound_flight_terminal = details_data['itineraries'][0]['segments'][0]['departure']['terminal']
                  except:
                    outbound_flight_terminal = 'Not yet assigned'

                  pattern = re.match(r'PT(\d+)H(\d+)M', outbound_flight['duration'])
                  if pattern:
                    hours = int(pattern.group(1))
                    minutes = int(pattern.group(2))
                    outbound_flight_duration = f'{hours}H {minutes}M'

                  # return flight data gathering
                  return_flight = details_data['itineraries'][1]
                  return_flight_duration = 'No data'
                  return_flight_terminal = 'Not yet assigned'
                  try:
                    return_flight_terminal = details_data['itineraries'][1]['segments'][0]['departure']['terminal']
                  except:
                    return_flight_terminal = 'Not yet assigned'

                  pattern = re.match(r'PT(\d+)H(\d+)M', return_flight['duration'])
                  if pattern:
                    hours = int(pattern.group(1))
                    minutes = int(pattern.group(2))
                    return_flight_duration = f'{hours}H {minutes}M'

                  try:
                    entry['Outbound Duration'] = outbound_flight_duration
                    entry['Outbound Terminal'] = outbound_flight_terminal
                    entry['Return Duration'] = return_flight_duration
                    entry['Return Terminal'] = return_flight_terminal
                    entry['Available Seats'] = seats
                    entry['Currency'] = currency
                  except:
                    print(f'{city} {entry} no details')

                else:
                  currency = 'No data'
                  seats = 'No data'

                  # outbound flight data gathering
                  outbound_flight = 'No data'
                  outbound_flight_duration = 'No data'
                  outbound_flight_terminal = 'No data'
                  # return flight data gathering
                  return_flight = 'No data'
                  return_flight_duration = 'No data'
                  return_flight_terminal = 'No data'

            data = pd.json_normalize(data)
            all_data[city] = data
        else:
            print(f"Error: {response.status_code}")
            print(response.json())

        # time.sleep(2)

    return all_data
access_token = get_token()
data = get_data(city_codes, '2024-12-16', access_token)
print(len(data))
```

```python
# make sure to replace NaN with 'Not yet assigned' if its a terminal or with 'No data' if anything else
access_token = get_token()

data['SYD'] = get_data(['SYD'], '2024-12-16', access_token)['SYD']
```

```python
data['SYD'] = data['SYD'].drop(['type', 'links.flightOffers', 'links.flightDates'], axis=1)
```

```python
for city in data:
  # data[city] = data[city].drop(['type', 'links.flightOffers', 'links.flightDates'], axis=1)
  data[city][['Outbound Duration','Return Duration','Available Seats','Currency']] = data[city][['Outbound Duration','Return Duration','Available Seats','Currency']].fillna(value='No data')
  data[city][['Outbound Terminal', 'Return Terminal']] = data[city][['Outbound Terminal', 'Return Terminal']].fillna(value='Not yet assigned')
```

```python
all_cities_data = pd.concat(data.values(), ignore_index=True)
```
